```python
import numpy as np
import matplotlib.pyplot as plt

# Definici√≥n de los puntos de interpolaci√≥n (datos del problema)
x_points = np.array([0.5, 1.0, 1.5, 2.0])  # Posici√≥n en metros
y_points = np.array([1.2, 2.3, 3.7, 5.2])   # Deformaci√≥n en mm

# Funci√≥n de interpolaci√≥n de Lagrange (sin cambios)
def lagrange_interpolation(x, x_points, y_points):
    n = len(x_points)
    result = 0
    for i in range(n):
        term = y_points[i]
        for j in range(n):
            if i != j:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        result += term
    return result

# a) Determinar la deformaci√≥n en x = 1.25 m
x_eval = 1.25
y_eval = lagrange_interpolation(x_eval, x_points, y_points)
print(f"a) La deformaci√≥n esperada en x = {x_eval} m es: {y_eval:.2f} mm")

# b) Graficar la interpolaci√≥n en el rango [0.5, 2.0]
x_values = np.linspace(min(x_points), max(x_points), 100)
y_values = [lagrange_interpolation(x, x_points, y_points) for x in x_values]

plt.figure(figsize=(8,5))
plt.plot(x_values, y_values, label="Interpolaci√≥n de Lagrange", color="blue")
plt.scatter(x_points, y_points, color="red", label="Datos experimentales")
plt.scatter([x_eval], [y_eval], color="green", label=f"Punto interpolado (x={x_eval}m)")
plt.xlabel("Posici√≥n en la viga (m)")
plt.ylabel("Deformaci√≥n (mm)")
plt.title("Deformaci√≥n de la viga - Interpolaci√≥n de Lagrange")
plt.legend()
plt.grid(True)
plt.savefig("deformacion_viga.png")
plt.show()
```
